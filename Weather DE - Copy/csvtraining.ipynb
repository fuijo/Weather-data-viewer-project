```python
import pandas as pd
```

```python
# Export the City_Data into a csv
city_data_df = pd.read_csv("Data_output/cities.csv")
northern_df = pd.read_csv("Data_output/northern_cities.csv")
southern_df = pd.read_csv("Data_output/southern_cities.csv")
```

```python
city_data_df
```

```python
def lower_columns(df):
    columns = df.columns.tolist()
    columns = [c.lower() for c in columns]
    df.columns = columns 
    df = df.reset_index().rename(columns={'index': 'id'})
    
    return df
```

```python
city_data_df = lower_columns(city_data_df)
city_data_df
```

```python
northern_df = lower_columns(northern_df)
northern_df
```

```python
southern_df = lower_columns(southern_df)
southern_df
```
